```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df['Gender']=pd.get_dummies(df['Gender'],drop_first=True)
```

```python
df['Gender']=df['Gender'].astype('int')
```

```python
df.head()
```

```python
df['Age'].unique()
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

```python
from sklearn import preprocessing
```

```python
label_encoder=preprocessing.LabelEncoder()
```

```python
df['Age']=label_encoder.fit_transform(df['Age'])
```

```python
df['Age'].unique()
```

```python
df.head()
```

```python
city_unique = sorted(df['City_Category'].dropna().unique())
city_mapping = {cat: idx for idx, cat in enumerate(city_unique)}
df['City_Category_Num'] = df['City_Category'].map(city_mapping).astype(int)
city_mapping
```

```python
if df['City_Category'].nunique() == 2:
    df_city = df[['City_Category_Num']].rename(columns={'City_Category_Num': 'City_Category_Binary'})
else:
    df_city = pd.get_dummies(df['City_Category'], prefix='City', drop_first=True, dtype=int)

df_city.head()
```

```python
df_city.head()
```

```python
df=pd.concat([df,df_city],axis=1)
```
